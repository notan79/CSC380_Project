```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./clean_data/updated_fully_merged_data.csv", dtype=np.float64)
df.head()
```

```python
need_norm = ["MEDHINC_CY", "WLTHINDXCY"]
norm = (df[need_norm] - df[need_norm].mean())/df[need_norm].std()
norm.mean(), norm.std(), norm.head()
```

```python
df[need_norm] = norm
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
train, test = train_test_split(df, train_size=0.8, test_size=0.2)
```

```python
train.head()
```

```python
test.head()
```

```python
lr = linear_model.LinearRegression()
```

```python
inp = ["sex", "age", "day", "month", "WLTHINDXCY"]
oup = ["fel_misd"]
lr.fit(train[inp], train[oup])
```

```python
lr.score(test[inp], test[oup])
```
